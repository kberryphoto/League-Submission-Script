```python
import pandas as pd
import numpy as np
```

```python
#Import your Datasheet
nerdData = "Nurgle_Datasheet.csv"

nerdData_df = pd.read_csv(nerdData, encoding="ISO-8859-1")
```

```python
#Import your Player Dictionary Tab
playerDict = "PlayerDict_25SEP19.csv"

playerDict_df = pd.read_csv(playerDict, encoding="ISO-8859-1")
```

```python
# Verify that the Data Sheet tab imported correctly
nerdData_df.head(5)
```

```python
#Clean up the Datasheet tab to only display information required for ITC Point Submission
cleanedNerd_df = nerdData_df[["ITC Finish", "Player Name", "Primary Faction", "Season", "Pod", "Score", "Score.1", "Score.2", "Score.3", "Score.4"]]
cleanedNerd_df.head(5)
```

```python
# Fill the NAs to prevent future errors
cleanedNerd_df["Score"].fillna(0, inplace=True)
cleanedNerd_df["Score.2"].fillna(0, inplace=True)
cleanedNerd_df["Score.3"].fillna(0, inplace=True)
cleanedNerd_df["Score.4"].fillna(0, inplace=True)
cleanedNerd_df["Score.1"].fillna(0, inplace=True)
cleanedNerd_df.head(5)
```

```python
#Begin Player Dictionary manipulation
cleanedPlayedDict_df = playerDict_df[["First", "Last", "Email Address"]]
cleanedPlayedDict_df.head(5)
```

```python
# Merge Dictionary Names together
cleanedPlayedDict_df['Player Name'] = cleanedPlayedDict_df['First'] + " " + cleanedPlayedDict_df['Last']
cleanedPlayedDict_df.head(5)
```

```python
#Merge the Data Sheet tab with the Player Dictionary tab
merge_table = pd.merge(cleanedNerd_df, cleanedPlayedDict_df, on="Player Name", how="inner")
merge_table.head(5)
```

```python
#Narrow the data to the League/Season of your choice.  First league is Season 1, count up sequentially afterwards
super_clean_nerd = merge_table.loc[merge_table["Season"] == 5, :]
super_clean_nerd.head(5)
```

```python
#Sort the merged table by Season, then Pod, then ITC Finish
sortedData = super_clean_nerd.sort_values(by=['Season', 'Pod', 'ITC Finish'])
sortedData.head(5)
```
